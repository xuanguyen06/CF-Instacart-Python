# Importing

```python
# Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

# Data Visualization Exercises

## Question 2

### Bar chart

```python
# Create bar chart of 'orders_day_of_week'
bar_dow = df['orders_day_of_week'].value_counts().plot.bar()

plt.title('Orders by Day of the Week', fontsize = 20)
plt.ylabel('Amount of Orders (in millions)')
plt.xlabel('Day of the Week')
```

### Histogram

```python
# Create histogram for 'prices'
hist_price = df['prices'].plot.hist(bins = 25)

plt.title('Frequencies of Prices', fontsize = 20)
plt.ylabel('Frequency (in millions)')
plt.xlabel('Price')
```

### Scatterplot

```python
# Create a scatterplot of 'prices'
scatplot_price = sns.scatterplot(x = 'prices', y = 'prices', data = df)

plt.title('Scatterplot of Prices', fontsize = 20)
```

### Create a sample subset

```python
# Create a sample subset of the dataframe using 70/30 ratio
np.random.seed(4)
dev = np.random.rand(len(df)) <= 0.7
```

```python
# Split dataframe into two samples - 70% in big, 30% in small
big = df[dev]
small = df[~dev]
```

```python
# Check results
len(df)
```

```python
len(big) + len(small)
```

### Line Chart

```python
# Reduce samples to only necessary columns
df_2 = small[['orders_day_of_week', 'prices']]
```

```python
# Create line chart
line_dow_price = sns.lineplot(data = df_2, x = 'orders_day_of_week', y = 'prices')

plt.title('Orders Day of the Week by Prices', fontsize = 20)
plt.xlabel('Day of the Week')
plt.ylabel('Prices')
```

# Histogram of Orders by Hour

## Question 3

```python
# Create a histogram of 'order_hour_of_day' utilizing 25 bins
hist_hour = df['order_hour_of_day'].plot.hist(bins = 100)

plt.title('Histogram of Orders by Hour', fontsize = 20)
plt.ylabel('Frequency (in millions)')
plt.xlabel('Hour of Day')
```

### The histogram shows the frequency of orders based on the hour of the day. Here we see that the peak hours are between 9am and 5pm. 

# Distribution of Loyalty Customers

## Question 4

```python
# Create a bar chart of 'loyalty_flag'
bar_loyalty = df['loyalty_flag'].value_counts().plot.bar(color = ['blue', 'green', 'red'])

plt.title('Distribution of Orders by Loyalty', fontsize = 20)
plt.ylabel('Frequency (in millions)', fontsize = 12)
plt.xlabel('Loyalty Flag', fontsize = 12)
plt.xticks(rotation = 0)
```

# Difference in Expenditure

## Question 5

```python
# Reduce the samples to the necessary columns
line_df = small[['prices', 'order_hour_of_day']]
```

```python
line_df.head()
```

```python
# Create a line chart based on the 'prices' and 'order_hour_of_day'
line_price_hour = sns.lineplot(data = line_df, x = 'order_hour_of_day', y = 'prices')

plt.title('Expenditure by Hour of Day', fontsize = 20)
plt.ylabel('Prices')
plt.xlabel('Hour of Day')
```

# Exploratory Analysis

## Question 6

```python
# Reduce samples to necessary columns
line_df_2 = small[['age', 'number_of_dependants']]
```

```python
line_df_2.head()
```

```python
# Create a line chart based on the age and number of dependents
line_age_dep = sns.lineplot(data = line_df_2, x = 'age', y = 'number_of_dependants')

plt.title('Line Chart of Age and Dependents', fontsize = 20)
plt.ylabel('Number of Dependents')
plt.xlabel('Age')
```

# Scatterplot of Age and Income

## Question 7

```python
# Create a scatterplot on age and income
scatplt_age_income = sns.scatterplot(x = 'age', y = 'income', data = df)

plt.title('Scatterplot of Age and Income', fontsize = 20)
plt.ylabel('Income')
plt.xlabel('Age')
```

### The scatterplot shows a positive correlation between age and income. The age can be grouped into those less than 40 yo and those above 40 yo. Those above 40 yo has a higher income, but the vast majority of customers in both groups make less than 200,000.
