## Dropping Columns

## Transposing Data

```python
# Transpose df_dep
df_dep.T
```

```python
# Overwrite df_dep with transposed df
df_dep_t = df_dep.T
```

```python
# Add an index to df_dep_t
df_dep_t.reset_index()
```

```python
# Create a new header by taking first row of df_dep_t
new_header = df_dep_t.iloc[0]
```

```python
new_header
```

```python
# Create a new dataframe to copies rows beyond the first row
df_dep_t_new = df_dep_t[1:]
```

```python
df_dep_t_new
```

```python
# Set header to dataframe
df_dep_t_new.columns = new_header
```
