# Creating Flags

## Question 4

```python
# Create a column for total orders of each customer
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
# Create the loyalty flag for 'Loyal customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Create the loyalty flag for 'Regular customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Create the loyalty flag for 'New customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

## Question 5

```python
# Check the spending habits of loyalty customer groups
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max', 'median', 'sum']})
```

## Question 6

```python
# Create a column for average order price for each customer
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
# Create spending flag for 'Low spender'
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spender_flag'] = 'Low spender'
```

```python
# Create spending flag for 'High spender'
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spender_flag'] = 'High spender'
```

```python
ords_prods_merge.head(15)
```

```python
# Check the frequency 
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

## Question 7

```python
# Create a column on the frequency of each customer
ords_prods_merge['med_freq'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# Create frequency flag for 'Non-frequent customer'
ords_prods_merge.loc[ords_prods_merge['med_freq'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
# Create frequency flag for 'Regular customer'
ords_prods_merge.loc[(ords_prods_merge['med_freq'] > 10) & (ords_prods_merge['med_freq'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
# Create frequncy flag for 'Frequent customer'
ords_prods_merge.loc[ords_prods_merge['med_freq'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.tail(20)
```

```python
# Check the frequency
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

### One thing to note is the 'NaN' values. This is attributed to the fact that the user made one single purchase and therefore there is no value associated with the 'days_since_prior_order'.
