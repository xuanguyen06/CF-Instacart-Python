## If-Statements

```python
# Create a subset of the first 1 million rows
df_subset = df[:1000000]
```

```python
# Define a function
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
# Apply the function
df_subset['price_range'] = df_subset.apply(price_label, axis = 1)
```

```python
df_subset['price_range'].value_counts(dropna = False)
```

```python
df_subset['prices'].max()
```

## If-Statements with loc()

### Subset dataframe

```python
df_subset.loc[df_subset['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
df_subset.loc[(df_subset['prices'] <= 15) & (df_subset['prices'] > 5), 'price_range_loc'] = 'Mid range product'
```

```python
df_subset.loc[df_subset['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

```python
df_subset['price_range_loc'].value_counts(dropna = False)
```

### Whole dataframe

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

## If-Statements with For-Loops

```python
# Frequency of orders by day of week
df['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create new column to store the busiest day
result = []

for value in df['orders_day_of_week']:
    if value == 0:
        result.append('Busiest Day')
    elif value == 4:
        result.append('Least Busy')
    else:
        result.append('Regularly Busy')
```

```python
result
```

```python
# Add list to dataframe
df['busiest_day'] = result
```

```python
df['busiest_day'].value_counts(dropna = False)
```

```python
# Create a new column to store new labels
result2 = []

for value in df['orders_day_of_week']:
    if (value == 0) | (value == 1):
        result2.append('Busiest days')
    elif (value == 3) | (value == 4):
        result2.append('Slowest days')
    else:
        result2.append('Regular days')
```

```python
# Add list to dataframe
df['busiest_day'] = result2
```

## Question 4

```python
# Check frequency of orders by hour of day
df['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Create a list to store the hour of day
result3 = []

for value in df['order_hour_of_day']:
    if (value >= 9) & (value <= 16):
        result3.append('Most orders')
    elif ((value >= 0) & (value <= 6)) | (value == 23):
        result3.append('Least orders')
    else:
        result3.append('Average orders')
```

```python
result3
```

```python
# Add list to df
df['busiest_period_of_day'] = result3
```
